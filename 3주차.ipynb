```python
temp = 20

if temp >=20:
    print("덥다")
elif temp >=10:
    print("따뜻하다")
elif temp >=0:
    print("춥다")
else:    
    print("매우 춥다")
```

반복문과 조건문 함께 사용

```python
scores = [42,60,77,89,20]

for s in scores:
    if s >= 60:
        print(f"{s}점 합격 입니다.")
else:   
    print(f"{s}점 불합격 입니다.")


```
